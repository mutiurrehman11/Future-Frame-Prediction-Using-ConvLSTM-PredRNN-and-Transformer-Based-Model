## Step 1: Preparing Data

```python
import pandas as pd

# File paths
train_csv_path = "train.csv"
val_csv_path = "val.csv"
test_csv_path = "test.csv"

# Load CSV files
train_data = pd.read_csv(train_csv_path)
val_data = pd.read_csv(val_csv_path)
test_data = pd.read_csv(test_csv_path)

# Inspect the data
print(train_data.head())
```

```python
# Remove leading slash for relative paths
train_data['clip_path'] = train_data['clip_path'].apply(lambda x: x.lstrip('/'))
val_data['clip_path'] = val_data['clip_path'].apply(lambda x: x.lstrip('/'))
test_data['clip_path'] = test_data['clip_path'].apply(lambda x: x.lstrip('/'))

# Verify changes
print(train_data['clip_path'].head())

print (val_data['clip_path'].head())
```

```python
import cv2
import numpy as np

def preprocess_video(video_path, target_size=(64, 64), grayscale=False):
    """
    Preprocess a single video by resizing frames and normalizing pixel values.
    """
    cap = cv2.VideoCapture(video_path)
    frames = []
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        # Resize frame
        frame = cv2.resize(frame, target_size)
        # Convert to grayscale if needed
        if grayscale:
            frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
            frame = frame[..., np.newaxis]  # Add channel dimension for grayscale
        frames.append(frame / 255.0)  # Normalize to [0, 1]
    cap.release()
    return np.array(frames)

sample_video_path = train_data.iloc[0]['clip_path']
frames = preprocess_video(sample_video_path)
print("Shape of preprocessed video:", frames.shape)  # (num_frames, 64, 64, channels)
```

```python
sequence_length = 20
target_length = 20
```

```python
import os
import numpy as np

# Process videos and save numpy arrays
def process_and_save_videos(data, output_dir):
    """
    Preprocess videos and save them as .npy files.
    Ensures the output directory exists for each file.
    """
    inputs = []
    targets = []
    for _, row in data.iterrows():
        video_path = row['clip_path']
        clip_name = row['clip_name']

        # Preprocess the video
        frames = preprocess_video(video_path, target_size=(64, 64), grayscale=False)
        if frames is not None:  # Only save if preprocessing was successful
            for i in range(0, (len(frames) - sequence_length - target_length), sequence_length + target_length):
                inputs.append(frames[i: i + sequence_length])  # Input sequence
                targets.append(frames[i + sequence_length: i + sequence_length + target_length])
            print(f"Processed and saved: {video_path}")
        else:
            print(f"Skipping video: {video_path} (could not preprocess)")

    # Ensure the output directory exists
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Save the processed videos as .npy files
    inputs_path = os.path.join(output_dir, 'inputs.npy')
    targets_path = os.path.join(output_dir, 'targets.npy')
    np.save(inputs_path, np.array(inputs))
    np.save(targets_path, np.array(targets))

    return np.array(inputs), np.array(targets)

# Use the current working directory as the output directory
output_dir = os.getcwd()  # Current directory where the script is located
val_inputs, val_targets = process_and_save_videos(val_data, os.path.join(output_dir, "val"))
test_inputs, test_targets = process_and_save_videos(test_data, os.path.join(output_dir, "test"))
```

```python
# Use the current working directory as the output directory
output_dir = os.getcwd()  # Current directory where the script is located

# Process the training videos
train_inputs, train_targets = process_and_save_videos(train_data, os.path.join(output_dir, "train"))
```

## Models

### PRED-RNN

```python
# Use the current working directory as the output directory
output_dir = os.getcwd()  # Current directory where the script is located

# Process the validation videos
val_inputs, val_targets = process_and_save_videos(val_data, os.path.join(output_dir, "val"))
```

```python
import tensorflow as tf
from tensorflow.keras.layers import ConvLSTM2D, Conv3D, MaxPooling3D, BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input
import os
import numpy as np

# Custom callback to print the epoch number
class PrintEpoch(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs=None):
        print(f"Epoch {epoch + 1}/{self.params['epochs']} - Loss: {logs['loss']:.4f} - Val Loss: {logs['val_loss']:.4f}")

def build_predrnn(input_shape):
    inputs = Input(shape=input_shape)
    
    # Add a ConvLSTM layer with a larger filter size to capture temporal dependencies
    x = ConvLSTM2D(filters=64, kernel_size=(3, 3), padding='same', return_sequences=True)(inputs)
    x = BatchNormalization()(x)
    
    # Add pooling layer (optional)
    x = MaxPooling3D(pool_size=(2, 2, 2))(x)  # Adjust pooling size if needed
    
    # Adding another ConvLSTM layer for further temporal processing
    x = ConvLSTM2D(filters=64, kernel_size=(3, 3), padding='same', return_sequences=True)(x)
    x = BatchNormalization()(x)
    
    # Upsample to increase resolution back to (64, 64)
    x = tf.keras.layers.UpSampling3D(size=(2, 2, 2))(x)  # Upsampling to match (64, 64) resolution
    
    # Output layer: Change the kernel size and number of frames to match the target shape
    x = Conv3D(filters=3, kernel_size=(3, 3, 3), activation='sigmoid', padding='same')(x)  # RGB output
    
    model = Model(inputs, x)
    return model


input_shape = (20, 64, 64, 3)  # Sequence length, height, width, channels (should match your data)


# Build the model
model = build_predrnn(input_shape)
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])

# Setup checkpoint and training history tracking
checkpoint_cb = tf.keras.callbacks.ModelCheckpoint(
    'predrnn_checkpoint.h5', save_best_only=True, monitor='val_loss', mode='min', verbose=1
)

history_cb = tf.keras.callbacks.History()
 

# Define the number of epochs
epochs = 20

# Use the processed data from previous steps
# Replace with actual paths or variables where you saved your data
train_data = np.array(train_inputs)  # Shape: (num_samples, seq_len, height, width, channels)
train_labels = np.array(train_targets)  # Shape: (num_samples, seq_len, height, width, channels)

val_data = np.array(val_inputs)  # Validation data
val_labels = np.array(val_targets)  # Validation labels

# Train the model with callbacks for checkpoint saving and history tracking
history = model.fit(
    train_data, train_labels, 
    epochs=epochs, 
    batch_size=8, 
    validation_data=(val_data, val_labels), 
    callbacks=[checkpoint_cb, history_cb, PrintEpoch()]
)

# Optionally save the final model
model.save("predrnn_model.h5")
```

```python
import matplotlib.pyplot as plt

# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend(['Train', 'Val'], loc='upper right')
plt.show() 

# Plot training & validation accuracy values
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(['Train', 'Val'], loc='upper left')
plt.show()
```

```python
import cv2
import numpy as np

def save_video(frames, output_path, fps=10):
    """
    Save frames as a video file.
    Args:
    - frames: Array of frames to save as a video (shape: num_frames, height, width, channels).
    - output_path: Path to save the video.
    - fps: Frames per second for the video.
    """
    # Ensure frames are in [0, 255] range and uint8 format
    frames = (frames * 255).astype(np.uint8)
    
    # Get video dimensions
    height, width, channels = frames[0].shape
    
    # Initialize video writer
    writer = cv2.VideoWriter(output_path, cv2.VideoWriter_fourcc(*'mp4v'), fps, (width, height))
    
    for frame in frames:
        # Convert RGB to BGR for OpenCV
        frame_bgr = cv2.cvtColor(frame, cv2.COLOR_RGB2BGR)
        writer.write(frame_bgr)
    
    writer.release()
    print(f"Video saved as: {output_path}")
```

```python
import matplotlib.pyplot as plt

def display_frames(input_frames, predicted_frames, num_rows=4):
    """
    Display input and predicted frames side by side.
    Args:
    - input_frames: NumPy array of input frames (sequence_length, height, width, channels).
    - predicted_frames: NumPy array of predicted frames (output_frames, height, width, channels).
    - num_rows: Number of rows to display the frames (default is 2: input on top, predicted below).
    """
    sequence_length = len(input_frames)
    output_frames = len(predicted_frames)

    # Total columns: max(sequence_length, output_frames)
    num_cols = 4
    fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5))

    # Display input frames in the first row
    for i in range(num_cols):
        if i < sequence_length:
            axes[0, i].imshow(input_frames[i])
            axes[0, i].set_title(f"Input Frame {i+1}")
        axes[0, i].axis("off")

    # Display predicted frames in the second row
    for i in range(num_cols):
        if i < output_frames:
            axes[1, i].imshow(predicted_frames[i])
            axes[1, i].set_title(f"Predicted Frame {i+1}")
        axes[1, i].axis("off")

    plt.tight_layout()
    plt.show()

# Call the function to display frames
display_frames(input_sequence, predicted_frames)
```
